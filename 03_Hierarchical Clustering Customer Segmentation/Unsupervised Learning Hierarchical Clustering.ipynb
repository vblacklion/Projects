## Introduction to Business Problem

Let's assume that we are working at a mall as a data scientist. Our aim is to create sales opportunities for the mall. In order to do that we want to build an unsupervised learning algorithm which is hierarchical Clustering using our Customers data. And we want to apply market segmentation on this group of customers to put them in distinct clusters based on common behaviors. At the end we will find the best cluster to target our sales offers.

## Loading and Exploring Dataset

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv('Customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

## Dendogram

How we can choose the optimal number of clusters? One and simple way to solve this problem is using a Dendrogram.

```python
# importing the library for dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Plotting Dendrogram using the columns 'Age','Salary' and 'Spendings'
plt.figure(figsize=(12,8))
X = df.loc[:, 'Age':'Spendings(1-100)'].values
dendrogram = dendrogram(linkage(X, method='ward')) # method='ward' is selected to minimize the variance of each cluster


plt.title('Customer Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidian Distance')
plt.show()
```

Now we can find the optimum number for Clustering our data. From above Dendrogram it is clear to see that setting Cluster size to 3 will be a good choice for our dataset.

### Hierarchical Clustering

```python
# importing libraries
from sklearn.cluster import AgglomerativeClustering as AC

# fitting our model with Hiearchical Clustering
hc = AC(n_clusters=3, affinity='euclidean', linkage='ward')

# predictions
y_hc = hc.fit_predict(X)
```

### Visualizing the Results

```python
plt.figure(figsize=(12,8))
plt.scatter(X[y_hc == 0, 1], X[y_hc == 0, 2], s=100, c='cyan', label='1st Cluster')
plt.scatter(X[y_hc == 1, 1], X[y_hc == 1, 2], s=100, c='green', label='2nd Cluster')
plt.scatter(X[y_hc == 2, 1], X[y_hc == 2, 2], s=100, c='red', label='3rd Cluster')

plt.title('Clusters of Customers (Annual Salary vs Spendings)')
plt.xlabel('Annual Salary $1000s')
plt.ylabel('Spendings (1-100)')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 2], s=100, c='cyan', label='1st Cluster')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 2], s=100, c='green', label='2nd Cluster')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 2], s=100, c='red', label='3rd Cluster')

plt.title('Clusters of Customers (Age vs Spendings)')
plt.xlabel('Age')
plt.ylabel('Spendings (1-100)')
plt.show()
```

### Conclusion

1. Customers with an Annual salary above $70000 can be grouped into two nearly equal categories. One groups spending very less, that means they prefer cheap or on sale products, while the other group spending 70-95 percent, that means they use our mall very often or they buy in big amounts.

2. Our analysis also shows an interesting result according to age groups. From the second Figure we see that age is an important factor in our customers. Customers, whose age between 27-40 are the ones that spends more in our mall with a percent of above 70.

3. According to these interesting results we can adjust our advertising strategies to:
    1. address customers especially over 40 to increase their spending.
    2. attract young customers (below 27) with promotions or with new products.
